# Populating SDM

## Import libraries

```python
import sqlite3
import pandas as pd
```

## Define connections

```python
path = "db"

accessoireVerkoopConn = sqlite3.connect(path + '/BikeToDrive_1_Accessoireverkoop.db')
fietsVerkoopConn = sqlite3.connect(path + '/BikeToDrive_2_Fietsverkoop.db')
onderhoudConn = sqlite3.connect(path + '/BikeToDrive_3_Onderhoud.db')
accessoireInkoopConn = sqlite3.connect(path + '/BikeToDrive_4_Accessoire_Inkoop.db')
fietsInkoopConn = sqlite3.connect(path + '/BikeToDrive_5_Fiets_Inkoop.db')
sdmConn = sqlite3.connect('BikeToDriveSDM.db')
```

## Accessoire verkoop

```python
# Filiaal
df_filiaal = pd.read_sql("SELECT * FROM Filiaal", accessoireVerkoopConn)
df_filiaal.to_sql('accessoire_verkoop_filiaal', sdmConn, if_exists='replace', index=False)

# Klant
df_klant = pd.read_sql("SELECT * FROM Klant", accessoireVerkoopConn)
df_klant.to_sql('accessoire_verkoop_klant', sdmConn, if_exists='replace', index=False)

# Monteur
df_monteur = pd.read_sql("SELECT * FROM Monteur", accessoireVerkoopConn)
df_monteur.to_sql('accessoire_verkoop_monteur', sdmConn, if_exists='replace', index=False)
# Leverancier
df_leverancier = pd.read_sql("SELECT * FROM Leverancier", accessoireVerkoopConn)
df_leverancier.to_sql('accessoire_verkoop_leverancier', sdmConn, if_exists='replace', index=False)

# Accessoires
df_accessoires = pd.read_sql("SELECT * FROM Accessoire", accessoireVerkoopConn)
df_accessoires.to_sql('accessoire_verkoop_accessoire', sdmConn, if_exists='replace', index=False)
# Accessoireverkoop
df_verkoop = pd.read_sql("SELECT * FROM Accessoire_Verkoop", accessoireVerkoopConn)
df_verkoop.to_sql('accessoire_verkoop_verkoop', sdmConn, if_exists='replace', index=False)
```

## Fiets verkoop

```python
# Klant
df_klant = pd.read_sql("SELECT * FROM Klant", fietsVerkoopConn)
df_klant.to_sql('fiets_verkoop_klant', sdmConn, if_exists='append', index=False)

# Filiaal
df_filiaal = pd.read_sql("SELECT * FROM Filiaal", fietsVerkoopConn)
df_filiaal.to_sql('fiets_verkoop_filiaal', sdmConn, if_exists='append', index=False)

# Monteur
df_monteur = pd.read_sql("SELECT * FROM Monteur", fietsVerkoopConn)
df_monteur.to_sql('fiets_verkoop_monteur', sdmConn, if_exists='append', index=False)

# Fiet
df_fiets = pd.read_sql("SELECT * FROM Fiets", fietsVerkoopConn)
df_fiets.to_sql('fiets_verkoop_fiets', sdmConn, if_exists='replace', index=False)

# Fiets verkopen
df_verkoop_fiets = pd.read_sql("SELECT * FROM Fiets_Verkoop", fietsVerkoopConn)
df_verkoop_fiets.to_sql('fiets_verkoop_verkoop', sdmConn, if_exists='replace', index=False)

# Fabrikant
df_fabrikant = pd.read_sql("SELECT * FROM Fabrikant", fietsVerkoopConn)
df_fabrikant.to_sql('fiets_verkoop_fabrikant', sdmConn, if_exists='replace', index=False)   
```

## Onderhoud

```python
# Filiaal
df_filiaal = pd.read_sql("SELECT * FROM Filiaal", onderhoudConn)
df_filiaal.to_sql('onderhoud_filiaal', sdmConn, if_exists='append', index=False)

# Monteur
df_monteur = pd.read_sql("SELECT * FROM Monteur", onderhoudConn)
df_monteur.to_sql('onderhoud_monteur', sdmConn, if_exists='append', index=False)

# Fiets
df_fiets = pd.read_sql("SELECT * FROM Fiets", onderhoudConn)
df_fiets.to_sql('onderhoud_fiets', sdmConn, if_exists='replace', index=False)

# Onderhoud
df_onderhoud = pd.read_sql("SELECT * FROM Onderhoud", onderhoudConn)
df_onderhoud.to_sql('onderhoud_onderhoud', sdmConn, if_exists='replace', index=False)

# Fabrikant
df_fabrikant = pd.read_sql("SELECT * FROM Fabrikant", onderhoudConn)
df_fabrikant.to_sql('onderhoud_fabrikant', sdmConn, if_exists='replace', index=False)
```

## Accessoire inkoop

```python
# Leverancier
df_leverancier = pd.read_sql("SELECT * FROM Leverancier", accessoireInkoopConn)
df_leverancier.to_sql('accessoire_inkoop_leverancier', sdmConn, if_exists='replace', index=False)

# Accessoires
df_accessoires = pd.read_sql("SELECT * FROM Accessoire", accessoireInkoopConn)
df_accessoires.to_sql('accessoire_inkoop_accessoire', sdmConn, if_exists='replace', index=False)
# Accessoireinkoop
df_accessoireinkoop = pd.read_sql("SELECT * FROM Accessoire_Inkoop", accessoireInkoopConn)
df_accessoireinkoop.to_sql('accessoire_inkoop_inkoop', sdmConn, if_exists='replace', index=False)
```

## Fiets inkoop

```python
# Fabrikant
df_fabrikant = pd.read_sql("SELECT * FROM Fabrikant", fietsInkoopConn)
df_fabrikant.to_sql('fiets_inkoop_fabrikant', sdmConn, if_exists='replace', index=False)

# Fiets
df_fiets = pd.read_sql("SELECT * FROM Fiets", fietsInkoopConn)
df_fiets.to_sql('fiets_inkoop_fiets', sdmConn, if_exists='replace', index=False)
# Fiets inkoop
df_inkoop_fiets = pd.read_sql("SELECT * FROM Fiets_Inkoop", fietsInkoopConn)
df_inkoop_fiets.to_sql('fiets_inkoop_inkoop', sdmConn, if_exists='replace', index=False)
```

## Close connections

```python
accessoireVerkoopConn.close()
fietsVerkoopConn.close()
onderhoudConn.close()
accessoireInkoopConn.close()
fietsInkoopConn.close()
sdmConn.close()
```
